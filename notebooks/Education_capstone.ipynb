```python
#avg percentage of income spent on non-essentials per region and education
#Region: to account for price differences between regions
percspending231avg=round(spending231.groupby(['Region', 'Education level'])['Non-Essential Spending (%)'].mean(),2)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=percspending231avg, x='Education level', y='Non-Essential Spending (%)', hue='Region', palette= 'viridis' )
plt.title('2023 Q1 Average Percentage of Non-Essential Spending by Education Level and Region')
plt.xlabel('Education Level')
plt.ylabel('Average Non-Essential Spending (%)')
plt.legend(title='Region')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
